```python
import json
import pandas as pd
import numpy as np
import os
import random
```

```python
aqg_baseline = json.load(open('..\Automatic-Question-Generator\AutomaticQuestionGenerator\AQG_Questions.json'))
t5_e2e_baseline = json.load(open('..\MultiTask-T5\T5_e2e_Questions'))
text_files = os.listdir('..\..\Data\Chunked_Transcripts')
text_for_ques = {}
for file in text_files:
    data = json.load(open('..\..\Data\Chunked_Transcripts\\' + file))
    text_data = []
    for i in data.keys():
        text_data.append(data[i]['text'])
    text_for_ques[file] = text_data
```

```python
text_for_ques
```

```python
human_eval_selected = []
for keys in t5_e2e_baseline.keys():
    subkeys_t5 = list(t5_e2e_baseline[keys].keys())
    subkeys_aqg = list(aqg_baseline[keys].keys())
    common_subkeys = list(set(subkeys_t5) & set(subkeys_aqg))
    # sample 2 from common subkeys
    subkeys = random.sample(common_subkeys, 2)
    mega_ls = []
    for i in subkeys:
        human_eval_selected.append([
            t5_e2e_baseline[keys][i],
            aqg_baseline[keys][i],
            text_for_ques[keys][int(i)]
        ])
```

```python
human_eval_selected
```

```python
df = pd.DataFrame(human_eval_selected, columns=['T5_e2e', 'AQG', 'Text'])
```

```python
# Reorder columns
df = df[['Text', 'T5_e2e', 'AQG']]
```

```python
# Save to csv
df.to_csv('human_eval_selected.csv', index=False)
```
