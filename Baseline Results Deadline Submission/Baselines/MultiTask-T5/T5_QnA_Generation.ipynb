## Custom Data

```python
import json
import os
paths = os.listdir('/content')
json_paths = []
for path in paths:
  if 'json' in path:
    json_paths.append(path)
data = {}
for jpath in json_paths:
  with open('/content/' + jpath) as f:
    jdata = json.load(f)
  data[jpath] = jdata
```

```python
questions = {}
```

```python
for fname in data.keys():
  questions[fname] = {}
  print("FILE: ", fname)
  transcript = data[fname]
  for val in transcript.keys():
    text = transcript[val]['text']
    print(text)
    print()
    # print("NLP QGen Output")
    # print(nlp_q_gen(text))
    # print()
    # print("NLP MultiTask QA QG Output")
    # print(nlp_multitask_qa_qg(text))
    # print()
    print("NLP E2E QG Output")
    q = nlp_e2e_qg(text)
    questions[fname][val] = q
    print()
  print(" ========= FILE ENDED =========")
```

```python
with open('/content/T5_e2e_Questions', 'w') as outfile:
    json.dump(questions, outfile)
```
